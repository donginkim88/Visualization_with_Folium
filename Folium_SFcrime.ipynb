```python
g = geocoder.google('San Francisco, California')
sf = g.latlng  # get lat, long
```

# Visualizing Crime Map in SF 

```python
map = folium.Map(location=sf, zoom_start=12)
marker_cluster = folium.MarkerCluster().add_to(map)

for index, row in crime_df[:1000].iterrows():  #iterrows() to read both index and row
    
    folium.Marker(
        location = [row['Y'], row['X']], # i[0] for index, i[1] for each row
        #location = [i[1]['Y'], i[1]['X']], # i[0] for index, i[1] for each row
    ).add_to(marker_cluster)

map
```
